```python
month_counts = df['Month'].value_counts()

month_counts.plot(kind='line')

plt.title('Monthly Trend')
plt.xlabel('Month')
plt.ylabel('Number of Orders')

plt.show()
```

```python
df['Year'] = df['OrderDate'].dt.year
Line = df['Year'].value_counts().sort_index()
Line.plot(kind='line')
plt.title('Yearly Trend')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.show()
```

```python
## Top Products
top_products = df['ProductKey'].value_counts().head(10)
top_products.plot(kind='bar')
plt.title('Top 10 Products')
plt.show()
```

```python
##Top Customers
top_customers = df['CustomerKey'].value_counts().head(10)
top_customers.plot(kind='bar')
plt.title('Top 10 Customers')
plt.xlabel('Customer Key')
plt.ylabel('Number of Orders')
plt.show()
```

```python
top_customers = df['TerritoryKey'].value_counts().head(10)
top_customers.plot(kind='bar')
plt.title('Top 10 Territories')
plt.xlabel('TerritoryKey')
plt.ylabel('Number of Orders')
plt.show()
```

```python
## Business Insights
## Orders by  year
year_orders = df.groupby('Year')['OrderNumber'].count()
print(year_orders)
```

```python
## Order by Month
df['Month'] = df['OrderDate'].dt.month_name()
month_orders = df.groupby('Month')['OrderNumber'].count()
print(month_orders)
```

```python
##Order By Territory
territory_orders = df.groupby('TerritoryKey')['OrderNumber'].count()
print(territory_orders.sort_values(ascending=False))
```

```python
## Top Products
product_orders = df.groupby('ProductKey')['OrderNumber'].count()
print(product_orders.sort_values(ascending=False).head(10))
```

```python
##Top Customers
customer_orders = df.groupby('CustomerKey')['OrderNumber'].count()
print(customer_orders.sort_values(ascending=False).head(10))
```
